```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
V = np.linspace(0, 1.21, 100)

Is = 2.92e-9  # saturation current in A
n = 1.92    # ideality factor
T = 293  # temperature in K
K  = 1.38e-23  # Boltzmann constant in J/K
q = 1.6e-19  # charge of an electron in C


I = Is * (np.exp(q * V / (n * K * T)) - 1)  # diode current equation

plt.plot(V, I, label='Diode I-V Characteristic (Linear)', color='blue')
plt.title('Diode I-V Characteristic')
plt.xlabel('Voltage (V)')
plt.ylabel('Current (mA)')

IM = [1, 5, 10, 20, 30, 40, 50, 60]  # current values in A
VM = [.61, .69, .72, .762, .78, .8, .82, .83]  # voltage values in V

plt.scatter(VM, IM, color='red', label='Measured Data')

plt.legend()
plt.grid(which='both', linestyle='--', linewidth=0.5)

plt.show()

plt.semilogy(V, I, label='Diode I-V Characteristic (Log)', color='blue')
plt.title('Diode I-V Characteristic')
plt.xlabel('Voltage (V)')
plt.ylabel('Current (mA)')

plt.scatter(VM, IM, color='red', label='Measured Data')

plt.legend()
plt.grid(which='both', linestyle='--', linewidth=0.5)

plt.show()
```
